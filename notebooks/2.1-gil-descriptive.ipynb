```python
import pandas as pd

# Load merged data
merged = pd.read_csv('../data/processed/merged_some_NaN.csv')
# show
merged.head()
```

```python
selected_cols = ["Dependency_Ratio", "Health_Expenditure", "RD_Expenditure", "Contribution_revenue", "Education_perGDP"]
merged[selected_cols].describe().T.round(2)
```

```python
import matplotlib.pyplot as plt
from pandas.plotting import table

desc_table = merged[selected_cols].describe().T.round(2)
# Create a figure
fig, ax = plt.subplots(figsize=(10, 4)) # Adjust width and height
ax.axis('off') # No axis
table(ax, desc_table, loc='center', cellLoc='center')

# Save to png
plt.savefig('../reports/figures/desc_stats.png', bbox_inches='tight', dpi=300)

# Show 
plt.show()
```
