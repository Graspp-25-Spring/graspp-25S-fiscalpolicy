## Health Expenditure vs Dependency Ratio 

### Health Expenditure vs Dependency Ratio (in total)

```python
import pandas as pd
import plotly.express as px
```

```python

# Load your data
df = pd.read_csv("../data/interim/merged_data.csv")

# Filter years 20002022
df_filtered = df[(df["Year"] >= 2000) & (df["Year"] <= 2022)].copy()

# Define custom color and order for income levels
income_colors = {
    "High income": "green",
    "Upper middle income": "purple",
    "Lower middle income": "yellow",
    "Low income": "red"
}
income_order = ["High income", "Upper middle income", "Lower middle income", "Low income"]

# Create animated scatter plot
fig = px.scatter(
    df_filtered,
    x="Dependency_Ratio",
    y="Health_Expenditure",
    animation_frame="Year",
    animation_group="Country",
    color="income_level",
    color_discrete_map=income_colors,
    category_orders={"income_level": income_order},
    hover_name="Country", 
    title="Health Expenditure vs. Dependency Ratio (20002022)",
    range_x=[df["Dependency_Ratio"].min() - 5, df["Dependency_Ratio"].max() + 5],
    range_y=[df["Health_Expenditure"].min() - 2, df["Health_Expenditure"].max() + 2]
)

# Style labels
fig.update_traces(textposition="top center", marker=dict(size=12))
fig.update_layout(legend_title_text="Income Level")

# Show the figure
fig.show()
```

## Corelation Visualization with Fitted Line

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Corelation of Health Expenditure and Dependency Ratio

```python
# Load the data first
df = pd.read_csv("../data/interim/merged_data.csv")

# Then define the custom color palette
custom_palette = {
    "High income": "green",
    "Upper middle income": "purple",
    "Lower middle income": "yellow",
    "Low income": "red"
}

# Set plotting style
sns.set(style="whitegrid")

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Make scatterplots + fitted lines
sns.scatterplot(x="Dependency_Ratio_Old", y="Health_Expenditure", hue="income_level", palette=custom_palette, data=df, ax=axes[0], alpha=0.6)
sns.regplot(x="Dependency_Ratio_Old", y="Health_Expenditure", data=df, ax=axes[0], scatter=False, color='black')
axes[0].set_title("Health Expenditure vs Dependency Ratio Old")

sns.scatterplot(x="Dependency_Ratio_Young", y="Health_Expenditure", hue="income_level", palette=custom_palette, data=df, ax=axes[1], alpha=0.6)
sns.regplot(x="Dependency_Ratio_Young", y="Health_Expenditure", data=df, ax=axes[1], scatter=False, color='black')
axes[1].set_title("Health Expenditure vs Dependency Ratio Young")

sns.scatterplot(x="log_GDP_percapita", y="Health_Expenditure", hue="income_level", palette=custom_palette, data=df, ax=axes[2], alpha=0.6)
sns.regplot(x="log_GDP_percapita", y="Health_Expenditure", data=df, ax=axes[2], scatter=False, color='black')
axes[2].set_title("Health Expenditure vs log GDP per Capita")

# Remove individual legends
for ax in axes:
    ax.legend_.remove()

# Add a single legend at the bottom
handles, labels = axes[2].get_legend_handles_labels()
fig.legend(handles, labels, title="Income Level", loc='lower center', ncol=4)

# Adjust layout and space for bottom legend
plt.tight_layout(rect=[0, 0.8, 1, 1])  
plt.subplots_adjust(bottom=0.25)

plt.show()
```

### Interpretation of the graphs

**Health Expenditure vs Dependency Ratio Old**
- High-income countries (green) show a strong upward trend. These countries tend to have both high elderly ratios and high health spending, reflecting advanced aging and stronger fiscal capacity.
- Upper-middle income countries (purple) cluster below the high-income group but still show a moderate increase in spending as elderly population grows.
- Lower-middle income (yellow) and low-income (red) countries have low old-age dependency and lower health expenditure, showing little to no increase in spending despite changes in aging.
- ***Conclusion:*** 
Only high-income countries can consistently increase health spending with aging, showing the unequal fiscal response capacity across income levels.

**Health Expenditure vs Dependency Ratio Young**
- Low-income (red) and lower-middle income (yellow) countries dominate the high youth dependency range, but spend the least on health, revealing constraints in providing adequate healthcare for growing young populations.
- High-income (green) and upper-middle income (purple) countries have lower youth dependency and higher health spending.
- Overall trend is downward, and low-income countries are the most strained.
- ***Conclusion:*** 
In countries with a high youth burden, especially low-income ones, health expenditure does not increase proportionally, showing a mismatch between demographic needs and fiscal response.

**Health Expenditure vs log GDP per Capita**
- High-income countries (green) cluster on the far right (high GDP per capita) and spend the most on health. Their upward trend is steep and consistent.
- Upper-middle income (purple) countries show a moderate increase.
- Lower-middle (yellow) and low-income (red) countries are spread on the left with lower spending and lower GDP.
- There's a clear income-based gradient.
- ***Conclusion:*** 
Health spending strongly increases with income level. High-income countries invest significantly more per capita in health services.

```python
import statsmodels.formula.api as smf

df = pd.read_csv("../data/interim/merged_data.csv")

# Filter data to years 20002022 and drop rows with missing values in key variables
df_model = df[
    (df["Year"] <= 2022) &
    df["Health_Expenditure"].notna() &
    df["Dependency_Ratio_Old"].notna() &
    df["Dependency_Ratio_Young"].notna() &
    df["log_GDP_percapita"].notna()
]

# Run OLS regression
model = smf.ols(
    formula="Health_Expenditure ~ Dependency_Ratio_Old + Dependency_Ratio_Young + log_GDP_percapita",
    data=df_model
).fit()

# Print regression summary
print(model.summary())
```

### Regression Interpretation

From the regression results, we find that the R-squared value of 0.488 indicates that the model explains 48.8% of the variation in health expenditure. The model's highly significant p-value (p < 0.001) confirms that the regression as a whole is statistically meaningful.

The regression coefficients allow us to interpret the individual relationships between each independent variable and health expenditure:
- A 1% increase in the old-age dependency ratio is associated with a 0.1732 unit increase in health expenditure.
- A 1% increase in the young-age dependency ratio leads to a 0.0421 unit increase in health expenditure.
- A 1% increase in GDP per capita (in log terms) corresponds to a 0.8212 unit increase in health expenditure.

To visualize these relationships, three scatter plots with regression lines are provided:
1. Health Expenditure vs. Old Dependency Ratio:
    This graph shows a clear positive correlation, consistent with the regression result. As the proportion of elderly individuals increases, health expenditure tends to rise. The upward trend supports the strong and significant coefficient in the regression.
2. Health Expenditure vs. Young Dependency Ratio:
    While the regression suggests a positive association, the scatter plot shows a visually negative trend. This discrepancy may indicate the presence of nonlinear relationships, confounding factors, or interaction effects. Additionally, the scatter appears more dispersed, suggesting lower explanatory power.
3. Health Expenditure vs. Log GDP per Capita:
    This plot displays a strong positive correlation, in line with the regression coefficient. It suggests that wealthier countries tend to allocate more resources to health, supporting the economic intuition that higher income levels enable greater public spending on health systems.

**Conclusion:**
The regression analysis reveals that health expenditure is significantly influenced by demographic structure and economic development. Specifically, countries with a higher old-age dependency ratio tend to spend more on health, reflecting the rising demand for medical care in aging populations. Although the young-age dependency ratio also shows a positive effect in the regression, the scatter plot suggests a more complex or weaker relationship, possibly due to unobserved factors. GDP per capita, meanwhile, exhibits a strong and clear positive correlationwealthier countries consistently allocate more resources to healthcare. Overall, the model explains 48.8% of the variation in health spending, underscoring that both demographic pressures and economic capacity are key drivers of national health expenditure patterns.
